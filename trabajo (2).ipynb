```python
# variables.py

# Dimensiones del tablero
TAMANO = 10

# Barcos del juego: nombre -> eslora (tamaño)
BARCOS = {
    "Barco1": 4,
    "Barco2": 3,
    "Barco3": 3,
    "Barco4": 2,
    "Barco5": 2,
    "Barco6": 2,
    "Barco7": 1,
    "Barco8": 1,
    "Barco9": 1,
    "Barco10": 1
}

# Símbolos para representar el tablero
AGUA = "·"
BARCO = "B"
TOCADO = "X"
FALLO = "O"



```

```python
# clases.py

class Tablero:
    def __init__(self, jugador_id):
        self.jugador_id = jugador_id
        self.tablero = np.full((TAMANO, TAMANO), AGUA)  # tablero con barcos
        self.tablero_visible = np.full((TAMANO, TAMANO), AGUA)  # lo que ve el rival
        self.barcos = BARCOS
        self.vidas = sum(self.barcos.values())
        self.colocar_barcos()

    def colocar_barcos(self):
        """Coloca los barcos de forma fija (puedes luego hacerlo aleatorio)."""
        posiciones = [
            (0, 0, 4, "H"),
            (2, 2, 3, "H"),
            (5, 5, 2, "V"),
            (9, 0, 1, "H"),
        ]
        for (x, y, eslora, orientacion) in posiciones:
            for i in range(eslora):
                if orientacion == "H":
                    self.tablero[x, y+i] = BARCO
                else:
                    self.tablero[x+i, y] = BARCO

    def disparar(self, x, y):
        """Gestiona un disparo sobre el tablero."""
        if self.tablero[x, y] == BARCO:
            self.tablero[x, y] = TOCADO
            self.vidas -= 1
            print("¡Tocado!")
            return True
        elif self.tablero[x, y] == AGUA:
            self.tablero[x, y] = FALLO
            print("Agua...")
            return False
        else:
            print("Ya habías disparado ahí.")
            return False

    def imprimir_tablero(self, visible=False):
        """Imprime el tablero. Si visible=True, muestra el tablero visible (sin barcos)."""
        print()
        if visible:
            print("Tablero visible:")
            print(self.tablero_visible)
        else:
            print("Tu tablero:")
            print(self.tablero)
        print()
        
```

```python
# funciones.py

import random
from variables import TAMANO

def pedir_coordenadas():
    """Pide coordenadas al jugador."""
    while True:
        try:
            x = int(input(f"Introduce la coordenada X (0-{TAMANO-1}): "))
            y = int(input(f"Introduce la coordenada Y (0-{TAMANO-1}): "))

            if 0 <= x < TAMANO and 0 <= y < TAMANO:
                return x, y
            else:
                print("Coordenadas fuera de rango. Intenta de nuevo.\n")

        except ValueError:
            print("Debes introducir números enteros.\n")


def coordenadas_aleatorias():
    """Genera coordenadas aleatorias para la máquina."""
    x = random.randint(0, TAMANO - 1)
    y = random.randint(0, TAMANO - 1)
    return x, y
    
    
```

```python
# main.py

turno_jugador = True

from clases import Tablero
from funciones import pedir_coordenadas, coordenadas_aleatorias

print("Bienvenido a HUNDIR LA FLOTA")
print("Intenta hundir todos los barcos enemigos antes de que la máquina hunda los tuyos.\n")

# Inicializar tableros
jugador = Tablero("Jugador")
maquina = Tablero("Máquina")

# Bucle principal
while jugador.vidas > 0 and maquina.vidas > 0:
    if turno_jugador:
        print("Tu turno:")
        maquina.imprimir_tablero(visible=True)
        x, y = pedir_coordenadas()
        acierto = maquina.disparar(x, y)
        turno_jugador = acierto  # sigue si acierta
    else:
        print("Turno de la máquina...")
        x, y = coordenadas_aleatorias()
        acierto = jugador.disparar(x, y)
        turno_jugador = not acierto

print("\n¡Juego terminado!")
if jugador.vidas == 0:
    print(":( Has perdido. La máquina ha hundido todos tus barcos.")
else:
    print(":(¡Has ganado! Hundiste todos los barcos enemigos.")
    
 
    
```
